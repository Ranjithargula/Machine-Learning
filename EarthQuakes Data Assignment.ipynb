```python
l1=[]
l2=[]
for i in df['time']:
    l1.append(i.split(' ')[0])
    l2.append(i.split(' ')[1])
df['Time']=l2
df['date']=l1
del df['time']
df.head()
```

```python
df=df.set_index('id')
df['mag']=df['mag'].fillna(df['mag'].interpolate())
df['mag']
df=df.sort_values(by='mag')[::-1]
df['place']
df['place']=df['place'].str.lstrip()
df.head()
```

```python
loc=[]
cntry=[]
for i in df['place']:
    if ',' in i:
        loc.append((i.split(',')[0]).lstrip())
        cntry.append((i.split(',')[1]).lstrip())
    else:
        cntry.append(i.lstrip())

df['country/state']=cntry
df.head()
```

```python
from collections import Counter
count=Counter(cntry)
dic=dict(count)
dic
```

```python
s=list(dic.values())
s.sort(reverse=True)
t=s[:11]
print(t)
```

```python
T=[]
for i in t:
    for j in dic:
        if i==dic[j]:
            T.append(j)
            print(j,'-',i)
```
